```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Atividade 1

```python
df = pd.read_csv('levenshtein_results.csv')
df
```

```python
df.sort_values('input_size', ascending=False)
```

```python
df2 = df.melt(id_vars=['input_size'])
df2
```

```python
x=df['input_size']
y1=df['brute_force_evaluations']
y2=df['dynamic_programming_evaluations']

ax = plt.scatter(x, y1, color='purple', alpha=0.4)
plt.scatter(x, y2, color='green', alpha=0.4)
plt.legend(['Força Bruta', 'Programação Dinâmica'])
plt.xlabel('Tamanho da Entrada (N*M)')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
sns.lineplot(x=df2['input_size'],
             y=df2['value'],
             hue=df2['variable'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df2,
             palette=['purple', 'green'],
)

plt.legend(['Força Bruta', 'Programação Dinâmica'])
plt.xlabel('Tamanho da Entrada (N*M)')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
ax = sns.lineplot(x=df['input_size'],
             y=df['brute_force_evaluations'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='purple'
)

bigo_y = [0.9 * (i ** 4) for i in df['input_size']]

sns.lineplot(x=df['input_size'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "0.9 x^4"])
plt.title('Força Bruta')
plt.xlabel('Tamanho da Entrada (N*M)')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
ax = sns.lineplot(x=df['input_size'],
             y=df['dynamic_programming_evaluations'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='green'
)

bigo_y = [10 + i for i in df['input_size']]

sns.lineplot(x=df['input_size'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "x + 10"])
plt.title('Programação Dinâmica')
plt.xlabel('Tamanho da Entrada (N*M)')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

# Atividade 2

```python
df = pd.read_csv('mss_results.csv')
df
```

```python
df.describe()
```

```python
df2 = df.melt(id_vars=['nodes_count', 'tree_height'])
df2
```

```python
x=df['nodes_count']
y1=df['divide_and_conquer_executions']
y2=df['greedy_executions']

ax = plt.scatter(x, y1, color='teal', alpha=0.4)
plt.scatter(x, y2, color='orange', alpha=0.4)
plt.legend(['Divisão & Conquista', 'Gulosa'])
plt.xlabel('Número de Nós')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
x=df['tree_height']
y1=df['divide_and_conquer_executions']
y2=df['greedy_executions']

ax = plt.scatter(x, y1, color='teal', alpha=0.4)
plt.scatter(x, y2, color='orange', alpha=0.4)
plt.legend(['Divisão & Conquista', 'Gulosa'])
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
x=df['tree_height']
y=df['greedy_executions']

sns.boxplot(x=x, y=y, color='orange')

plt.title('Paradigma Guloso')
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')
plt.show()
```

```python
x=df['tree_height']
y=df['divide_and_conquer_executions']

sns.boxplot(x=x, y=y, color='teal')

plt.title('Paradigma Divisão & Conquista')
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')
plt.show()
```

```python
sns.lineplot(x=df2['tree_height'],
             y=df2['value'],
             hue=df2['variable'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df2,
             palette=['orange', 'teal'],
)

plt.legend(['Guloso', 'Divisão & Conquista'])
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
x=df2['tree_height']
y=df2['value']

ax = sns.scatterplot(
    x=x,
    y=y,
    hue=df2['variable'],
    palette=['orange', 'teal'],
    alpha=0.4,
)

handles, labels  =  ax.get_legend_handles_labels()
ax.legend(handles, ['Gulosa', 'Divisão & Conquista'], loc='upper left')

plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')
plt.show()
```

```python
x=df2['nodes_count']
y=df2['value']

ax = sns.scatterplot(
    x=x,
    y=y,
    hue=df2['variable'],
    palette=['orange', 'teal'],
    alpha=0.4,
)

handles, labels  =  ax.get_legend_handles_labels()
ax.legend(handles, ['Gulosa', 'Divisão & Conquista'], loc='upper left')

plt.xlabel('Quantidade de Nós')
plt.ylabel('Iterações / Chamadas')
plt.show()
```

```python
ax = sns.lineplot(x=df['tree_height'],
             y=df['divide_and_conquer_executions'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='teal'
)

bigo_y = [6.25e-2 * (2 ** i) for i in df['tree_height']]

sns.lineplot(x=df['tree_height'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "6.25e-2 2ᴺ"])

plt.title('Divisão & Conquista')
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
ax = sns.lineplot(x=df['nodes_count'],
             y=df['divide_and_conquer_executions'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='teal'
)

bigo_y = [i + 5000 for i in df['nodes_count']]

sns.lineplot(x=df['nodes_count'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "x + 5000"])

plt.title('Divisão & Conquista')
plt.xlabel('Quantidade de Nós')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
import math

ax = sns.lineplot(x=df['tree_height'],
             y=df['greedy_executions'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='orange'
)

bigo_y = [5 * math.log(i) for i in df['tree_height']]

sns.lineplot(x=df['tree_height'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "5 log(x)"])

plt.title('Greedy')
plt.xlabel('Altura da Árvore')
plt.ylabel('Iterações / Chamadas')

plt.show()
```

```python
import math

ax = sns.lineplot(x=df['nodes_count'],
             y=df['greedy_executions'],
             err_style="band",
             ci='sd',
             estimator=np.mean,
             data=df,
             color='orange'
)

bigo_y = [8 * math.log(math.log(i)) for i in df['nodes_count']]

sns.lineplot(x=df['nodes_count'],
             y=bigo_y,
             err_style="band",
             ci='sd',
             estimator=np.mean,
             linestyle="--",
             color='gray'
)

plt.legend(["Observado", "8 log(log(x))"])

plt.title('Greedy')
plt.xlabel('Quantidade de Nós')
plt.ylabel('Iterações / Chamadas')

plt.show()
```
